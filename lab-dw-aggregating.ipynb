```python
# your code goes here
# Data Inspection Script
# url: https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')
print(df.head())
print(df.shape)
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.columns)
print(df.dtypes)
print(df.nunique())

# print unique values for each column
for col in df.columns:
    print(f"\nUnique values in column {col}: {df[col].unique()}")


```

```python
# Data Cleaning
# -----------------------
# Drop Unnamed: 0
# Standardize column names

df = df.drop(columns=["Unnamed: 0"], errors="ignore")

df.columns = (
    df.columns
      .str.lower()
      .str.replace(" ", "_", regex=False)
)

# Quick check
print(df.shape)
df.head()
```

```python
df["number_of_open_complaints"].dtype
```

At this stage, the dataset has appropriate data types and standardized categorical values to proceed with filtering and aggregation tasks.
>
Minor inconsistencies (e.g., NaN values or float representation of count data) do not affect the current analysis.



```python
# Task 1: Filter: low total_claim_amount + response "Yes"
# Create a new DataFrame with customers who:
# - Have low total_claim_amount (below 1000)
# - Responded "Yes" to the last marketing campaign


filtered_df = df[
    (df["total_claim_amount"] < 1000) &
    (df["response"] == "Yes")
]

filtered_df.head()
filtered_df.shape

print(filtered_df.shape)
print(filtered_df.head())

# save filtered_df to a csv file
filtered_df.to_csv("filtered_customers.csv", index=False)

# Filter customers who responded "Yes" and have total_claim_amount below 1000 (low-claim responders)
```

```python
# Task 2: Analyze average monthly premium and/or CLV by policy_type & gender (Response == "Yes") + compare with claims
# Using the original Dataframe, analyze:
# - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
# - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

yes_responders = df[df["response"] == "Yes"]
grouped_analysis = (
    yes_responders
    .groupby(["policy_type", "gender"], as_index=False)
    .agg({
        "monthly_premium_auto": "mean",
        "customer_lifetime_value": "mean",
        "total_claim_amount": "mean"
    })
    .round(2)
)


print(grouped_analysis.sort_values("customer_lifetime_value", ascending=False))
```

```python
# Task 3: Analyzeustomers per state, keep only states with > 500 customers
# Using the original Dataframe, analyze the total number of customers who have policies in each state, 
# and then filter the results to only include states where there are more than 500 customers.

state_customer_counts = df['state'].value_counts()
filtered_states = state_customer_counts[state_customer_counts > 500]
print(filtered_states)
```

**Insights:**
>
California has the highest number of customers, followed by Oregon and Arizona.
>
All states in the dataset have more than 500 customers, indicating a strong geographic coverage across the analyzed regions.

```python
# Task 4: Analyze Min, Max, Median CLV by education level and gender
# Using the original Dataframe, find the maximum, minimum, and median customer lifetime value 
# by education level and gender. 
# Write your conclusions.
clv_stats = (
    df
    .groupby(["education", "gender"], as_index=False)
    .agg({      
        "customer_lifetime_value": ["min", "max", "median"]
    })
)
clv_stats.columns = ["education_level", "gender", "min_clv", "max_clv", "median_clv"]
print(clv_stats.sort_values("median_clv", ascending=False))         
print(clv_stats.sort_values("median_clv", ascending=False))
```

**Insights:**
>
Customers with "High School or Below" education show the highest median CLV for both genders, especially males, indicating strong long-term value despite lower formal education levels.
>
Across most education levels, male customers tend to have slightly higher median CLV than females, although the differences are not large.
>
Doctorate-level customers show comparatively lower median CLV, suggesting that higher education does not necessarily translate into higher customer lifetime value in this dataset.
>
Overall, education level appears to influence CLV, but gender differences are moderate and should be interpreted with caution.

```python
# Bonus 5:
# Analyze the number of policies sold by state and month. 
# Present the data in a table where the months are arranged as columns and the states are arranged as rows.
# Policies sold by State (rows) and Month (columns)

# Parse the date column
df["effective_to_date"] = pd.to_datetime(df["effective_to_date"], errors="coerce")

# Create a month column (month number)
df["month"] = df["effective_to_date"].dt.month


# Sort months in ascending order
policy_counts = df.pivot_table(
    index="state",
    columns="month",
    aggfunc="size",
    fill_value=0
).sort_index(axis=1)

# show the result in descending order of total policies sold
policy_counts["total_policies"] = policy_counts.sum(axis=1)
policy_counts = policy_counts.sort_values("total_policies", ascending=False)
policy_counts = policy_counts.drop(columns=["total_policies"])


print(policy_counts)

# display the pivot table
policy_counts
```

**Insights:**
>
Policy sales are highly concentrated in California and Oregon across both months analyzed.
>
California leads consistently, making it the most important state in terms of sales volume.
>
Arizona also shows solid performance, while Nevada and Washington lag behind.
>
These patterns suggest that sales efforts and marketing resources may be most effective when prioritized in high-volume states, particularly at the beginning of the year.

```python
# Bonus 6:
# Find top 3 states by total policies sold,
# then show the number of policies sold by month for those states.

total_policies_by_state = (
    df.groupby("state")["policy"]
      .count()
      .sort_values(ascending=False)
)

top3_states = total_policies_by_state.head(3).index
print("\nTop 3 states by policies sold:", list(top3_states))

top3_policies_by_month = (
    df[df["state"].isin(top3_states)]
      .groupby(["state", "month"])["policy"]
      .count()
      .unstack(fill_value=0)
)
# show the result in descending order of total policies sold
top3_policies_by_month["total_policies"] = top3_policies_by_month.sum(axis=1)
top3_policies_by_month = top3_policies_by_month.sort_values("total_policies", ascending=False)
top3_policies_by_month = top3_policies_by_month.drop(columns=["total_policies"])

print("\nPolicies sold by Month for Top 3 states:")
print(top3_policies_by_month)

# display the pivot table
top3_policies_by_month
```

**Insights:**
>
California, Oregon, and Arizona are the top three states with the highest number of policies sold in both January and February. California leads by a significant margin, followed by Oregon and Arizona, showing consistent sales performance across months.
>
These states represent the strongest markets and should be prioritized for marketing and sales efforts.
>
The analysis is limited to the first two months of the year, so seasonal effects may not be fully captured.

```python

# Bonus 7:
# Marketing wants the response rate ("Yes") by sales channel.
# Steps:
# - Create a binary flag (1 if Yes else 0)
# - Compute mean of the flag by sales_channel (mean = response rate)

df["response_yes"] = (df["response"] == "Yes").astype(int)

response_rate_by_channel = (
    df.groupby("sales_channel")["response_yes"]
      .mean()
      .mul(100)
      .round(2)
      .sort_values(ascending=False)
)

print("\nResponse rate (%) by Sales Channel:")
print(response_rate_by_channel)
```

**Insights:**
>
The Agent channel shows the highest response rate (18.01%), significantly outperforming all other sales channels. Web, Branch, and Call Center have similar and considerably lower response rates (around 10â€“11%).
>
This suggests that personalized, agent-driven interactions are more effective at converting customers in marketing campaigns.
>
Results are based on response rate only and do not account for revenue, costs, or long-term value.
